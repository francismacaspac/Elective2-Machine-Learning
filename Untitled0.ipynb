```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv("Housing.csv")
# To display the top 5 rows 
df.head(5) 
```

```python
df.tail(5)   
```

```python
df.dtypes
```

```python
df = df.drop(['airconditioning', 'hotwaterheating'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"bedrooms": "BR", "stories": "STR", "mainroad": "MR"})
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count() 
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
df = df.dropna()  
df.count()
```

```python
print(df.isnull().sum()) 
```

```python
sns.boxplot(x=df['BR'])
```

```python
sns.boxplot(x=df['STR'])
```

```python
sns.boxplot(x=df['area'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df.price.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by price")
plt.ylabel('Number of cars')
plt.xlabel('price');
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['parking'], df['area'])
ax.set_xlabel('parking')
ax.set_ylabel('area')
plt.show()
```
